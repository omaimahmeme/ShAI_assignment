```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/Salaries.csv")
```

```python
data.head()
```

```python
# Convert 'Id' and 'Year' to object type
data['Id'] = data['Id'].astype(str)
data['Year'] = data['Year'].astype(str)

# Verify the changes
print(data.dtypes)
```

Basic Data Exploration: Identify the number of rows and columns in the dataset, determine the data types of each column, and check for missing values in each column.

```python
data.info()
```

```python
# Drop 'Notes' and 'Status' columns
data = data.drop(['Notes', 'Status'], axis=1)
```

Descriptive Statistics: Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation.

```python
basic_stats = data.describe()
print("\nBasic Statistics:")
print(basic_stats)

# Calculate mode
mode_salary = data['TotalPay'].mode()[0]
print("\nMode Salary:", mode_salary)

# Calculate range
range_salary = data['TotalPay'].max() - data['TotalPay'].min()
print("\nRange of Salaries:", range_salary)

# Calculate standard deviation
std_dev_salary = data['TotalPay'].std()
print("\nStandard Deviation of Salaries:", std_dev_salary)
```

Data Cleaning: Handle missing data by suitable method with explain why you use it.

```python
# Impute 'BasePay' with the mean
data['BasePay'].fillna(data['BasePay'].mean(), inplace=True)


# Impute 'OvertimePay' with the mean
data['OvertimePay'].fillna(data['OvertimePay'].mean(), inplace=True)

# Impute 'OtherPay' with the mean
data['OtherPay'].fillna(data['OtherPay'].mean(), inplace=True)

# Impute 'Benefits' with the mean
data['Benefits'].fillna(data['Benefits'].mean(), inplace=True)
```

```python
data.info()
```

Basic Data Visualization: Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.


```python
import matplotlib.pyplot as plt
import seaborn as sns


# Create a histogram for 'TotalPay'
plt.figure(figsize=(10, 6))
sns.histplot(data['TotalPay'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()


```

```python
# Create a bar chart for the proportion of employees in different departments
plt.figure(figsize=(12, 8))
department_counts = data['JobTitle'].value_counts().head(10)  # Top 10 departments
department_counts.plot(kind='bar', color='lightcoral')
plt.title('Proportion of Employees in Different Departments (Top 10)')
plt.xlabel('Job Title')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
salary_stats_by_job = data.groupby('JobTitle')['TotalPay'].agg(['mean', 'median', 'min', 'max', 'count'])

# Print the summary statistics
print(salary_stats_by_job)

# Compare average salaries across different job titles
average_salary_comparison = salary_stats_by_job.sort_values(by='mean', ascending=False)
print("\nAverage Salaries across Different Job Titles:")
print(average_salary_comparison)
```

Simple Correlation Analysis: Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.

```python
# Calculate correlation coefficients
correlation_matrix = data[['TotalPay', 'BasePay']].corr()

# Print correlation coefficients
print("Correlation Coefficients:")
print(correlation_matrix)

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BasePay', y='TotalPay', data=data, color='skyblue')
plt.title('Scatter Plot of BasePay vs TotalPay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.show()
```
